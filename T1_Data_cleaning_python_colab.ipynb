### Leer datos desde una URL

```python
def downloadFromURL(url, filename, sep = ",", delim = "\n", encoding="utf-8"):
    #primero importar la libreria y hacemos conexion con la web de los datos
    #señora de alta alcurnia de grandes apellidos subio al micro de la era donde dice que cobran cupos
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    print("el estado de la respuesta es %d" %(r.status))
    response = r.data
    
    #el objeto response contiene un string binario asi que lo convertimos a un string descodificandolo en UTF-8
    str_data = response.decode(encoding)
    #dividimos el string en un array de filas separandolo en intros
    lines = str_data.plit(delim)
    #la primera línea contiene la cabecera, así que la extraemos
    col_names = lines[0].split(sep)
    n_cols = len(col_names)
    #Generamos un diccionario vacío donde ira la informacion procesada desde la URL externa
    counter = 0
    main_dic = {}
    for col in col_name:
        main_dict[col] = []
    #Procesamos fila a fila la información para ir rellenando el diccionario con los datos como hicimos antes
    for line in lines:
        #Nos saltamos la primera linea que es la que contiene la cabecera y la tenemos procesada
        if(counter > 0):
            #Dividimos cada string por las comas como elemento separador
            values =  line.strip().split(sep)
            #Añadimos cada valor a su respectiva columna del diccionario
            for i in range(len(col_name)):
                main_dict[col_name[i]].append(values[i])
        counter += 1
    print("El data set tiene %d filas y %d columnas"%(counter, n_cols))
    #convertimos el diccionario procesado a Dataframe y comprobamos que los datos son correctos
    df = pd.DataFrame(main_dict)
    print(df.head())
    #Elegimos donde guardarlo (en la carpeta datasets es donde tiene mas sentido por el contexto del analisis)
    mainpath = r"C:\Users\HOME\Desktop\ML\python-ml-course-master\datasets"
    filename = "work\download"
    fullpath = os.path.join(mainpath, filename)
    
    #lo guaramos en csv, en JSON o en excel según queramos
    df.to_csv(fullpath+".csv")
    df.to_json(fullpath+".json")
    df.to_excel(fullpath+".xls")
    
    return
```
