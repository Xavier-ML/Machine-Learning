### Caso 1

Jill diseña paneles solares como hobby.

El 1 de abril, el diseño "Mark I" de Jill empieza a generar energía: 1 kJ/día.

El 1 de mayo, su diseño "Mark II" empieza a generar 4 kJ de energía al día.

¿Qué día es cuando el diseño Mark II de Jill ha generado tanta energía total como el diseño Mark I?
¿Cuánta energía total han generado ambos ese día?
¿Cuáles serían las soluciones a (1.) y (2). si el diseño Mark II generara 1kJ de energía al día?

**comenzamos con el primer panel solar**

El 1 de abril, el diseño "Mark I" de Jill empieza a generar energía: 1 kJ/día.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Desde el 1 de abril hasta el 1 de mayo son 30 dias transcurridos.

```python
t = np.linspace(0, 60, 100) # inicio, final, n puntos
```

Diseño "Mark I" de Jill empieza a generar energía 1 kJ/día, t medido en dias: $E = 1kJ/dia * t$

```python
E_1 = 1 * t
```

Diseño "Mark II" de Jill empieza a generar energía 4 kJ/día, t medido en dias: $E = 4kJ/dia * (t - 30)$

```python
E_2 = 4 * (t - 30)
```

```python
fig, ax = plt.subplots()
plt.title('paneles solares')
plt.xlabel('tiempo (en dias)')
plt.ylabel('energía (en kJ)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 60])
ax.plot(t, E_1, c='green')
ax.plot(t, E_2, c='red')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```

Diseño "Mark I" de Jill empieza a generar energía 1 kJ/día, t medido en dias: $E = 1kJ/dia * t$.

Está representa la recta verde.

Diseño "Mark II" de Jill empieza a generar energía 4 kJ/día, t medido en dias:  $E=4kJ/dia∗(t−30)$.

Está representa la recta roja.

¿Qué día es cuando el diseño Mark II de Jill ha generado tanta energía total como el diseño Mark I?

Se obtiene a los 40 dias donde se obtiene igual cantidad de energia.

¿Cuánta energía total han generado ambos ese día?

El diseño "Mark I" de Jill ha generado 40 kJ hasta los 40 dias, mientras que el diseño "Mark II" de Jill ha generado 40 kJ hasta los 40 dias, con un total de 80kJ hasta los 40 dias.

### Caso 2

¿Cuáles serían las soluciones a (1.) y (2). si el diseño Mark II generara 1kJ de energía al día?

Diseño "Mark I" de Jill empieza a generar energía 1 kJ/día, t medido en dias:  E=1kJ/dia∗t

```python
E_1_2 = 1 * t
```

Diseño "Mark II" de Jill empieza a generar energía 4 kJ/día, t medido en dias:  E=1kJ/dia∗(t−30)

```python
E_2_2 = 1* (t - 30)
```

```python
fig, ax = plt.subplots()
plt.title('paneles solares')
plt.xlabel('tiempo (en dias)')
plt.ylabel('energía (en kJ)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 60])
ax.plot(t, E_1_2, c='green')
ax.plot(t, E_2_2, c='blue')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
_ = plt.axhline(y=10, color='purple', linestyle='--')
```

¿Qué día es cuando el diseño Mark II de Jill ha generado tanta energía total como el diseño Mark I?

No hay un dia en el cual el diseño Mark II y el diseño Mark I, lleguen a tener la misma energia generada debido a que el diseño II genera la misma energia por dia, nunca alcalzaria al diseño I, que tiene una ventaga de 30 dias que se traduce en 30kJ  de energía.

¿Cuánta energía total han generado ambos ese día?

El diseño "Mark I" de Jill ha generado 40 kJ hasta los 40 dias, mientras que el diseño "Mark II" de Jill ha generado 10 kJ hasta los 40 dias debido al retrazo del tiempo de 30 dias, con un total de 50kJ hasta los 40 dias contados desde el 1 de Abril.
